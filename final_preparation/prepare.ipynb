```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("dnabert_ready_full_taxonomy.csv")
df
```

```python
db = pd.read_csv("dataset_animalia.csv")
db
```

```python

# Step 1: Get counts for all Primer_Family values (including NaN)
# Count NaN separately
nan_counts = df[df['Primer_Family'].isna()].groupby('Species').size()

# Count non-NaN primers
counts = (df[df['Primer_Family'].notna()]
          .groupby(['Species', 'Primer_Family'])
          .size()
          .unstack(fill_value=0))

# Step 2: Add 80% of NaN counts to the comparison
if 'folmer' in counts.columns and 'leray' in counts.columns:
    # Add 80% of window samples to each primer count
    # Convert to float to avoid dtype warning
    counts_adjusted = counts.astype(float)
    for species in counts.index:
        window_count = nan_counts.get(species, 0)
        weighted_window = window_count * 0.8
        counts_adjusted.loc[species, 'folmer'] += weighted_window
        counts_adjusted.loc[species, 'leray'] += weighted_window
    
    # Find species with both primers
    both_primers = counts[(counts['folmer'] > 0) & (counts['leray'] > 0)]
    
    # Determine which to remove based on adjusted counts
    # Remove leray if folmer >= leray (handles ties), else remove folmer
    remove_primer = np.where(
        counts_adjusted.loc[both_primers.index, 'folmer'] >= counts_adjusted.loc[both_primers.index, 'leray'],
        'leray', 
        'folmer'
    )
    
    to_remove = pd.DataFrame({
        'Species': both_primers.index,
        'Primer_Family': remove_primer
    })
    
    # Mark rows to remove using merge
    df['_remove'] = False
    df = df.merge(to_remove, on=['Species', 'Primer_Family'], how='left', indicator='_merge')
    df['_remove'] = df['_merge'] == 'both'
    df = df.drop('_merge', axis=1)
else:
    df['_remove'] = False

# Step 3: Keep NaN rows (no removal based on having non-NaN primers)

# Step 4: Filter
df_filtered = df[~df['_remove']].drop('_remove', axis=1)
```

```python
# Count each Primer_Family type
folmer_count = (df_filtered['Primer_Family'] == 'folmer').sum()
leray_count = (df_filtered['Primer_Family'] == 'leray').sum()
window_count = df_filtered['Primer_Family'].isna().sum()

# Calculate total
total = len(df_filtered)

# Calculate percentages
folmer_pct = (folmer_count / total * 100) if total > 0 else 0
leray_pct = (leray_count / total * 100) if total > 0 else 0
window_pct = (window_count / total * 100) if total > 0 else 0

# Print results
print(f"Folmer: {folmer_count:,} ({folmer_pct:.2f}%)")
print(f"Leray: {leray_count:,} ({leray_pct:.2f}%)")
print(f"Window (NaN): {window_count:,} ({window_pct:.2f}%)")
print(f"\nTotal: {total:,} (100.00%)")

"""
This will output something like:

Folmer: 450,123 (45.01%)
Leray: 234,567 (23.46%)
Window (NaN): 315,310 (31.53%)

Total: 1,000,000 (100.00%)
"""
```

```python
# Compare original database with filtered dataframe
samples_lost = len(db) - len(df_filtered)
print(f"Samples lost: {samples_lost} ({samples_lost/len(db)*100:.2f}%)")
print(f"Original samples: {len(db)}")
print(f"Remaining samples: {len(df_filtered)} ({len(df_filtered)/len(db)*100:.2f}%)")
```

```python
# Get species from original database
original_species = set(db['Species'].values)

# Get current species
current_species = set(df['Species'].values)

# Find missing species (in original but not in current)
missing_species = original_species - current_species

print(f"\n{'='*50}")
print(f"Species comparison:")
print(f"Original species count: {len(original_species)}")
print(f"Current species count: {len(current_species)} ({len(current_species)/len(original_species)*100:.2f}%)")
print(f"Missing species count: {len(missing_species)} ({len(missing_species)/len(original_species)*100:.2f}%)")
if missing_species:
    print(f"\nSpecies that are gone:")
    for species in sorted(missing_species):
        print(f"  - {species}")
else:
    print("\nNo species are missing!")
```

```python
# Create a dictionary mapping species to their taxonomic information
# This removes duplicates and is much more memory efficient
taxonomy_dict = db[['Species', 'Kingdom', 'Phylum', 'Class', 'Order', 'Family', 'Genus']].drop_duplicates(subset='Species').set_index('Species').to_dict('index')

print(f"Created taxonomy dictionary with {len(taxonomy_dict)} unique species")

# Map the taxonomic information to df
df['Kingdom'] = df['Species'].map(lambda x: taxonomy_dict.get(x, {}).get('Kingdom', None))
df['Phylum'] = df['Species'].map(lambda x: taxonomy_dict.get(x, {}).get('Phylum', None))
df['Class'] = df['Species'].map(lambda x: taxonomy_dict.get(x, {}).get('Class', None))
df['Order'] = df['Species'].map(lambda x: taxonomy_dict.get(x, {}).get('Order', None))
df['Family'] = df['Species'].map(lambda x: taxonomy_dict.get(x, {}).get('Family', None))
df['Genus'] = df['Species'].map(lambda x: taxonomy_dict.get(x, {}).get('Genus', None))

# Reorder columns to match the desired structure
column_order = [
    'Header',
    'DNABERT_Sequence',
    'Kingdom',
    'Phylum',
    'Class',
    'Order',
    'Family',
    'Genus',
    'Species'
]

final_df = df[column_order]

print("Taxonomy mapping complete!")
print(f"Final df shape: {df.shape}")
print("\nChecking for any missing taxonomy (NaN values):")
print(df[['Kingdom', 'Phylum', 'Class', 'Order', 'Family', 'Genus']].isna().sum())

final_df
```

```python
# Count occurrences of each genus
genus_counts = df['Genus'].value_counts()

# Define intervals and count classes in each
intervals = {
    '1 sample': (genus_counts == 1).sum(),
    '2-5 samples': ((genus_counts >= 2) & (genus_counts <= 5)).sum(),
    '6-10 samples': ((genus_counts >= 6) & (genus_counts <= 10)).sum(),
    '11-50 samples': ((genus_counts >= 11) & (genus_counts <= 50)).sum(),
    '51-100 samples': ((genus_counts >= 51) & (genus_counts <= 100)).sum(),
    '>100 samples': (genus_counts > 100).sum()
}
```

```python
# Print summary statistics
total_classes = len(genus_counts)
print("\nSummary:")
print(f"Total number of unique classes: {total_classes}")
print(f"Total number of samples: {len(df)}")
print("\nClasses per interval:")
for interval, count in intervals.items():
    percentage = (count / total_classes * 100) if total_classes > 0 else 0
    print(f"  {interval}: {count} classes ({percentage:.2f}%)")
```

```python
# Create bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(intervals.keys(), intervals.values(), color='steelblue', edgecolor='black')

# Calculate total classes for percentage
total_classes = len(genus_counts)

# Add value labels with percentages on top of bars
for bar in bars:
    height = bar.get_height()
    percentage = (height / total_classes * 100) if total_classes > 0 else 0
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}\n({percentage:.1f}%)',
             ha='center', va='bottom', fontsize=10)

plt.xlabel('Sample Count Interval', fontsize=12)
plt.ylabel('Number of Classes', fontsize=12)
plt.title('Distribution of Classes by Sample Count', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, max(intervals.values()) * 1.15)  # Add 15% padding at the top
plt.tight_layout()
plt.grid(axis='y', alpha=0.3)

plt.show()
```

```python
# Create bar chart with improved aesthetics
plt.figure(figsize=(14, 7))
plt.style.use('seaborn-v0_8-darkgrid')

# Create gradient colors for bars
colors = plt.cm.viridis(range(len(intervals)))

bars = plt.bar(intervals.keys(), intervals.values(), color=colors, edgecolor='white', linewidth=1.5, alpha=0.85)

# Calculate total classes for percentage
total_classes = len(genus_counts)

# Add value labels with percentages on top of bars
for bar in bars:
    height = bar.get_height()
    percentage = (height / total_classes * 100) if total_classes > 0 else 0
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}\n({percentage:.1f}%)',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.xlabel('Sample Count Interval', fontsize=13, fontweight='bold')
plt.ylabel('Number of Classes', fontsize=13, fontweight='bold')
plt.title('Distribution of Classes by Sample Count', fontsize=16, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=11)
plt.ylim(0, max(intervals.values()) * 1.15)
plt.tight_layout()
plt.grid(axis='y', alpha=0.4, linestyle='--', linewidth=0.7)

plt.show()
```
