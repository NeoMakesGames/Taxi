```python
taxonomy = "Species"
taxonomy_types = db[taxonomy].unique()
print(taxonomy_types)
print(len(taxonomy_types))
```

```python
values_to_exclude = taxonomy_types
classes_df_negative = db[~db[taxonomy].isin(values_to_exclude)]
#classes_df_negative
```

```python
values_to_include = taxonomy_types
classes_df_positive = db[db[taxonomy].isin(values_to_include)]
#classes_df_positive
```

```python
values = db['Sequence']
lengths = values.str.len()
all_same_length = lengths.nunique() == 1

if all_same_length:
    print(f"All strings have the same length: {lengths.iloc[0]}")
else:
    print("Strings have different lengths")
    print(lengths.value_counts())
    num_different_sizes = lengths.nunique()
    print(f"Number of different sizes: {num_different_sizes}")
```

```python
db['length'] = db['Sequence'].str.len()
class_counts = db['length'].value_counts()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(class_counts.values, bins=50, edgecolor='black')
plt.title('Distribution of Class Frequencies')
plt.xlabel('Number of Occurrences')
plt.ylabel('Number of Classes')
plt.yscale('log') 
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.hist(filtered_counts.values, bins=100, edgecolor='black', alpha=0.7)
plt.title('Distribution of Classes with 1-100 Occurrences')
plt.xlabel('Number of Occurrences')
plt.ylabel('Number of Classes')
plt.xlim(0, 100)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
