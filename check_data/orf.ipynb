```python
import pandas as pd
```

```python
df = pd.read_csv("coi_class_balance_checkpoint_leray.csv")
df
```

```python
db = pd.read_csv("dataset_animalia.csv")
db
```

```python
total_species = len(df)
print(f"Total number of species: {total_species}")
```

```python
x = 1  
species_below_threshold = df[df['Sample_Count'] <= x]
count_below = len(species_below_threshold)

print(f"\nNumber of species with less than {x} samples: {count_below}")
print(f"Percentage: {(count_below/total_species)*100:.2f}%")
```

```python
total_remaining_samples = df['Sample_Count'].sum()

samples_lost = len(db) - total_remaining_samples
print(f"Samples lost: {samples_lost} ({samples_lost/len(db)*100:.2f}%)")
print(f"Original samples: {len(db)}")
print(f"Remaining samples: {total_remaining_samples} ({total_remaining_samples/len(db)*100:.2f}%)")
```

```python
# Load species from text file
with open('species_list_filtered.txt', 'r') as f:
    original_species = set(line.strip() for line in f)

# Get current species
current_species = set(df['Species'].values)

# Find missing species (in original but not in current)
missing_species = original_species - current_species
print(f"\n{'='*50}")
print(f"Species comparison:")
print(f"Original species count: {len(original_species)}")
print(f"Current species count: {len(current_species)} ({len(current_species)/len(original_species)*100:.2f}%)")
print(f"Missing species count: {len(missing_species)} ({len(missing_species)/len(original_species)*100:.2f}%)")
if missing_species:
    print(f"\nSpecies that are gone:")
    for species in sorted(missing_species):
        print(f"  - {species}")
else:
    print("\nNo species are missing!")
```
